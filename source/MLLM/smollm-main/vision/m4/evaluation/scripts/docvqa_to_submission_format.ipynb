```python
import json
import ast

# Conditions to match
target_model_name = "/scratch/m4/tr_317_vsmollm_long_context/opt_step-1000/unwrapped_model"
target_image_size = 1920
target_dataset_split = "test"

last_matching_score = None  # Variable to store the score of the last matching row

# Open and read the JSONL file
with open(docvqa_test_results, 'r') as f:
    for i, line in enumerate(f):
        # Parse each line as a JSON object
        json_obj = json.loads(line.strip())
        
        # Check if the conditions match
        if (json_obj.get("model_name_or_path") == target_model_name and
            json_obj.get("image_size_after_scaling") == target_image_size and
            json_obj.get("dataset_split") == target_dataset_split):
            
            # Store the score of the current matching row
            last_matching_score = json_obj.get("score")

# After iterating through the file, print the last matching score
if last_matching_score is not None:
    print(f"Score from the last matching row: {last_matching_score}")
else:
    print("No matching rows found.")


# Convert the content to a Python dictionary using ast.literal_eval
# ast.literal_eval can safely evaluate the single-quote string into a Python dictionary
python_dict = ast.literal_eval(last_matching_score)

# Convert to a JSON-compliant format (optional, if you want to manipulate as JSON string)
json_data = json.dumps(python_dict)

# Load it as a JSON object
json_object = json.loads(json_data)

# Function to rename 'question_id' to 'questionId' and convert the id to integer
def modify_json(data):
    modified_data = []
    for item in data:
        # Rename 'question_id' to 'questionId' and convert it to an integer
        new_item = {
            'questionId': int(item['question_id']),  # Convert to integer
            'answer': item['answer']
        }
        modified_data.append(new_item)
    return modified_data

# Modify the json_object
modified_json_object = modify_json(json_object['server_results'])

a = open("CVC_submission.json","w")
a.write(json.dumps(modified_json_object))
a.close()
```
